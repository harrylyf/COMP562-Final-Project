```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
res50_pt = pd.read_csv('data/res50 pretrained/loss.csv', names=['iter', 'loss'])
res50_npt = pd.read_csv('data/res50 not pretrained/loss.csv', names=['iter', 'loss'])
# google_pt
# squeeze_pt
# squeeze_npt


```

```python
cnt = 0
tmp = 0
aaa = 0
res50_iter = []
res50_loss = []
for row in res50_pt.iterrows():
    tmp += row[1]['loss']
    cnt += 1
    if(cnt == 20):
        res50_loss.append(tmp/20)
        res50_iter.append(aaa)
        cnt = 0
        tmp = 0
        aaa += 1
```

```python
cnt = 0
tmp = 0
aaa = 0
res50npt_iter = []
res50npt_loss = []
for row in res50_npt.iterrows():
    tmp += row[1]['loss']
    cnt += 1
    if(cnt == 20):
        res50npt_loss.append(tmp/20)
        res50npt_iter.append(aaa)
        cnt = 0
        tmp = 0
        aaa += 1
```

```python
fig = plt.figure(dpi=1200)

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.plot(res50_iter, res50_loss, label="pre-trained model")
plt.plot(res50npt_iter, res50npt_loss, label='Without pre-trained')
# plt.title('title')
plt.ylabel('training loss')
plt.xlabel('iteration')

plt.legend()
plt.show()
plt.savefig('test.png')
```

```python
def get_accuracy(origin_data):
    _iter = []
    _loss = []
    for row in origin_data.iterrows():
        _iter.append(row[1]['iter'])
        _loss.append(row[1]['acc'])
    return _iter, _loss
```

```python
res50_npt = pd.read_csv('data/res50 not pretrained/loss.csv', names=['iter', 'loss'])
google_pt = pd.read_csv('data/google pretrained/loss.csv', names=['iter', 'loss'])
squeeze_pt = pd.read_csv('data/squeeze pretrained/loss.csv', names=['iter', 'loss'])
squeeze_npt = pd.read_csv('data/squeeze not pretrained/loss.csv', names=['iter', 'loss'])
spp_cnn = pd.read_csv('data/spp+cnn/loss.csv', names=['iter', 'loss'])
dense_pt = pd.read_csv('data/dense pretrained/loss.csv', names=['iter', 'loss'])

```

```python
def modify(origin_data):
    cnt = 0
    tmp = 0
    aaa = 0
    iter_ = []
    loss_ = []
    for row in origin_data.iterrows():
        tmp += row[1]['loss']
        cnt += 1
        if(cnt == 20):
            loss_.append(tmp/20)
            iter_.append(aaa)
            cnt = 0
            tmp = 0
            aaa += 1
    return iter_, loss_
```

```python
res50_npt_iter, res50_npt_loss = modify(res50_npt)
google_pt_iter, google_pt_loss = modify(google_pt)
squeeze_pt_iter, squeeze_pt_loss = modify(squeeze_pt)
squeeze_npt_iter, squeeze_npt_loss = modify(squeeze_npt)
spp_cnn_iter, spp_cnn_loss = modify(spp_cnn)
dense_pt_iter, dense_pt_loss = modify(dense_pt)
```

```python
fig = plt.figure(dpi=1200)
# fig = plt.figure()

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.plot(res50_npt_iter[3:400], res50_npt_loss[3:400], label="ResNet 50")
plt.plot(google_pt_iter[3:400], google_pt_loss[3:400], label='GoogleNet')
plt.plot(squeeze_pt_iter[5:400], squeeze_pt_loss[5:400], label='SqueezeNet')
# plt.plot(squeeze_npt_iter[:40], squeeze_npt_loss[:40], label='SqueezeNet2')
plt.plot(spp_cnn_iter[5:400], spp_cnn_loss[5:400], label='CNN')
plt.plot(dense_pt_iter[3:400], dense_pt_loss[3:400], label='DenseNet')


plt.ylabel('training loss')
plt.xlabel('iteration')

plt.legend()
plt.show()
```
